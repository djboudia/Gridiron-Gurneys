```python

def GetPreGameInjuries(year):
    info = []
    gametype = ["REG" + str(week) for week in range(1,19)] + ['POST1','POST2','POST3','PRO1','POST4']
    for week in gametype:
        time.sleep(1)
        url = f'https://www.nfl.com/injuries/league/{year}/{week}'
        try:
            res = requests.get(url)
            soup = BeautifulSoup(res.content, 'html.parser')
            headers = soup.find_all('h2',class_='d3-o-section-title')
            for date in headers:
                gamedate =  date.get_text()
                lst_els = date.find_next_siblings(attrs={"class": "nfl-o-injury-report__unit"}) 
                for elm in lst_els:
                    test = elm.findChildren('div', class_='d3-o-section-sub-title')
                    for team in test:
                        team_name = team.get_text()
                        # players = team.find_next_siblings()
                        players = elm.findChildren('div',class_="d3-o-table--horizontal-scroll")
                        for player in players[0].table.tbody:
                            info = [cell.get_text().lstrip() for cell in player]
                            info.append(team_name)
                            info.append(gamedate)
                            info.append(year)
                            yield info
                            info = ""
        except Exception as e:
            continue

  
            
    #     print(team)
    # for team in tbl:
        #     for row in team.tbody:
                # cells = [cell.get_text().lstrip() for cell in row]
                # if cells[1] in ['RB','QB','WR','TE']:
                #     data.append(cells)
```

```python
lst_df = []
for i in tqdm([2002,2003,2004,2005,2006,2007,2008,2009,2011,2012,2013]): 
    df = pd.DataFrame(data=GetPreGameInjuries(i),
                      columns=['player','position',
                               'injury','practice_status',
                               'game_status','team','gamedate','year'])
    lst_df.append(df)
```

```python

def GetMonth(x):
    
    if x == 'september':
        return 9
    elif x == 'october':
        return 10
    elif x == 'november':
        return 11
    elif x == 'december':
        return 12
    elif x == 'january':
        return 1
    elif x == 'february':
        return 2
    elif x == 'march':
        return 3
    elif x == 'august':
        return 8


months = { 'AUGUST': 8,
          'SEPTEMBER': 9,
          'OCTOBER': 10,
          'NOVEMBER': 11,
          'DECEMBER': 12,
          'JANUARY': 1,
          'FEBRUARY': 2,
          'MARCH': 3}



nfl['month'] = nfl['gamedate'].apply(lambda x: months[re.search(r', ([A-Za-z]+)', x).group(1)])
```
